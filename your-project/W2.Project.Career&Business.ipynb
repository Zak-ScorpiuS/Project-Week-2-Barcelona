```python
avg_salaries_industry = pd.read_csv('/Users/anna/Desktop/Ironhack/Week_2/Project-Week-2-Barcelona/avg_salaries_industry.csv', sep=';')
avg_salaries_industry 
```

```python
avg_salary_tech = pd.DataFrame(avg_salaries_industry, columns=['year','Información y comunicaciones'])
avg_salary_tech.rename(columns = {'Información y comunicaciones':"tech",},
                    inplace = True)
avg_salary_tech
```

```python
overall_salary_raw = pd.read_csv('/Users/anna/Desktop/Ironhack/Week_2/Project-Week-2-Barcelona/overall_salary.csv', sep = ";")
overall_salary = pd.DataFrame(overall_salary_raw, columns=['year','overall'])

overall_salary
```

```python
salary_comparison = pd.merge(overall_salary,avg_salary_tech, on='year')

salary_comparison ["ratio"] = salary_comparison["tech"]/ salary_comparison["overall"]

salary_comparison
```

```python
avg_overall_salary = salary_comparison['overall'].mean()

print("The average yearly salary of Cataluña from 2013-2017 is: ", avg_overall_salary)

avg_tech_salary = salary_comparison['tech'].mean()

print("The average yearly salary of Cataluña from 2013-2017 is: ", avg_tech_salary)

ratio_salaries = avg_tech_salary / avg_overall_salary 


print("The Tech salaries are on average ", "%.2f" % ratio_salaries, "bigger than the overall salary.")


```

```python
itservices_vs_industry = bcn_companies_evolution.loc[:,['year', 'IT_Services',"Industry"]].copy()
```

```python
itservices_vs_industry.groupby("year").sum()
#answer to 1st question, we can see how the IT Services sector has almost dobled the number
#of companies over the last 10 years
```

```python
# Company growth rate in the IT Industry

IT_2009=itservices_vs_industry.iloc[0,1]

IT_2018=itservices_vs_industry.iloc[9,1]

growth_rate_IT = ((IT_2018-IT_2009)/IT_2009)

print("In 2009, there were: ", IT_2009,"companies registered. Whereas in 2018 there was : ", IT_2018, "companies registered.")
print("The growth registered between both periods is of: ",round(growth_rate_IT*100,2),"%.")
```

```python
# Company growth rate of the Overall Industry

IND_2009=itservices_vs_industry.iloc[0,2]

IND_2018=itservices_vs_industry.iloc[9,2]

growth_rate_IT = (IND_2018-IND_2009)/IND_2009

print("In 2009, there were: ", IND_2009,"companies registered. Whereas in 2018 there was : ", IND_2018, "companies registered.")
print(f"The growth registered between both periods is of: {round(growth_rate_IT*100,2)}%.")
```

```python
itservices_vs_industry["% of total industry"] = (itservices_vs_industry ["IT_Services"]/itservices_vs_industry ["Industry"])*100

print(itservices_vs_industry)
```
