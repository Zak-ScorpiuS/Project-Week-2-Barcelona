```python
student_data['Class_Tech'] = student_data.Degree.apply(lambda x: True if x in tech_degree else False)
```

```python
# stem_degrees1 = student_data[student_data['Degree'].isin(tech_degree)]
```

```python
#stem_degrees1.shape
```

```python
stem_degrees = student_data[student_data.Class_Tech == True]
```

```python
#stem_degrees.shape
```

```python
stem_degrees.dtypes
```

```python
stem_degrees.Students = stem_degrees.Students.astype(int).copy()
```

```python
stem_year_evolution = stem_degrees.groupby('Year').sum()
```

```python
stem_year_evolution['Students_YoY'] = stem_year_evolution.Students.shift(1)
stem_year_evolution.iloc[0, 2] = 0
```

```python
stem_year_evolution
```

```python
stem_year_evolution['Students_YoY'] = stem_year_evolution[['Students', 'Students_YoY']]\
                .apply(lambda x: (x[0] - x[1]) /  x[1] if x[0] != 421 else 0, axis=1)
```

```python
stem_year_evolution = stem_year_evolution[['Students', 'Students_YoY', 'Class_Tech']]
```

```python
# The below graph showcases the growth on "Tech" related degrees, as well as the volume of Degrees and 
#Studies per year. On top of that we can see the Yearly growth rates in student volume.

stem_year_evolution
```

```python
mean_students = stem_year_evolution['Students'].mean()
mean_students_yoy = stem_year_evolution['Students_YoY'].mean()

print("On average, there are ", int(mean_students), "on `Tech degrees`on a yearly basis in Catalan universities")
print("The yearly growth of the volume of students averages ", int(mean_students_yoy*100),"%.")
```
